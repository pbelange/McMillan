```python
import subprocess
import sys
import os
import matplotlib.pyplot as plt
import numpy as np

from cpymad.madx import Madx
import xtrack as xt
import xpart as xp


# Importing LHC sequences:
with open('madx_input.log', 'w') as f:
    mad = Madx(stdout=f)
    mad.option(echo = True, warn = True)
    mad.call(file='000_makeSequence.mad')
    
    mad.command.beam(particle='proton',sequence='racetrack',charge=1,npart=1,energy=7000,bv=1)

    mad.input('use,sequence=racetrack')



# Build Xtrack line importing MAD-X expressions
line = xt.Line.from_madx_sequence(mad.sequence['racetrack'],
                                  deferred_expressions=True # <--
                                  )
# Define reference particle
line.particle_ref = xp.Particles(mass0=xp.PROTON_MASS_EV, q0=1,
                                 gamma0=mad.sequence.racetrack.beam.gamma)

# Build tracker
tracker = line.build_tracker()


tw  = tracker.twiss(method='4d').to_pandas()
srv = tracker.survey().to_pandas()





# %matplotlib widget

plt.figure()
plt.suptitle('Racetrack:survey')
plt.plot(srv['Z'],srv['X'],'-o')
plt.legend()
plt.axis('equal')
plt.xlabel('z (m)')
plt.ylabel(r'x (m)')
plt.axis('equal')

plt.figure(figsize=(15,5))
plt.plot(tw['s'],tw['betx'],'-',label='betx')
plt.plot(tw['s'],tw['bety'],'-',label='bety')
plt.plot(tw['s'],tw['dx'],'-',label='Dx')
plt.legend()


plt.figure(figsize=(15,5))
plt.plot(tw['s'],tw['alfx'],'-',label='alfx')
plt.plot(tw['s'],tw['alfy'],'-',label='alfy')
plt.legend()
```

```python
import ducktrack as dtk
import sympy as sp



x,px = sp.symbols('x p_x',real=True)
x_map = dtk.TestParticles(  x  = x,
                            px = px)

# Case to choose dtk class
from_xtrack = { 'Multipole':dtk.Multipole,
                'DriftExact':dtk.DriftExact,
                'DipoleEdge':dtk.DipoleEdge,
                'Drift':dtk.Drift}


for ee in line.elements:
    ee_dict = ee.to_dict()
    
    # Skip markers
    if ee_dict['__class__'] == 'Marker':
        continue
    
    # Copy as ducktrack element
    ee_dtk = from_xtrack[ee_dict['__class__']].from_dict(ee_dict)

    # Apply map
    ee_dtk.track(x_map) 

    # Truncate to 4th order:
    O_n = 8
    x_map.x  = (x_map.x.expand() + sp.O(x**O_n)+sp.O(px**O_n)).removeO()
    x_map.px = (x_map.px.expand() + sp.O(x**O_n)+sp.O(px**O_n)).removeO()
    x_map.y,x_map.py,x_map.zeta,x_map.delta = 0,0,0,0


print(40*'_')
display(x_map.x)
display(x_map.px)
```

```python
class OneTurnMap:
    def __init__(self, x_map,px_map):
        self.x_map  = sp.lambdify((x,px),x_map)
        self.px_map  = sp.lambdify((x,px),px_map)

    def iterate(self,x0,px0,n_turns):
        x,px = np.nan*np.ones(n_turns),np.nan*np.ones(n_turns)
        x[0],px[0] = x0,px0
        for ii in range(n_turns-1):
            x[ii+1],px[ii+1] = self.x_map(x[ii],px[ii]),self.px_map(x[ii],px[ii])
        return x,px
    



for O_map in [3,4,5,6,7,8]:


    testMap = OneTurnMap(   (x_map.x.expand() + sp.O(x**O_map)+sp.O(px**O_map)).removeO(),
                            (x_map.px.expand() + sp.O(x**O_map)+sp.O(px**O_map)).removeO())
    
    # Plotting resulting phase space

    plt.figure()
    plt.suptitle(f'Truncation order: {O_map}')
    particles,coordinates = generate_particles()
    
    plane = 'x'

    for part_x,part_px in zip(particles.x,particles.px):
        _x,_px =  testMap.iterate(part_x,part_px,1000)

        plt.plot(_x,_px,'o',color='k',alpha=0.5,ms=1,label='space')
            
        

    plt.axis('equal')
    plt.xlabel(f'{plane}_n')
    plt.ylabel(f'p{plane}_n')

    plt.xlim([-2e-4,2e-4])
    plt.ylim([-2e-4,2e-4])


```
