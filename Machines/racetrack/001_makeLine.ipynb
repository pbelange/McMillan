```python
import ducktrack as dtk
import sympy as sp



x,px = sp.symbols('x p_x',real=True)
x_map = dtk.TestParticles(  x  = x,
                            px = px)

# Case to choose dtk class
from_xtrack = { 'Multipole':dtk.Multipole,
                'DriftExact':dtk.DriftExact,
                'DipoleEdge':dtk.DipoleEdge,
                'Drift':dtk.Drift}


for ee in line.elements:
    ee_dict = ee.to_dict()
    
    # Skip markers
    if ee_dict['__class__'] == 'Marker':
        continue
    
    # Copy as ducktrack element
    ee_dtk = from_xtrack[ee_dict['__class__']].from_dict(ee_dict)

    # Apply map
    ee_dtk.track(x_map) 

    # Truncate to 4th order:
    O_n = 8
    x_map.x  = (x_map.x.expand() + sp.O(x**O_n)+sp.O(px**O_n)).removeO()
    x_map.px = (x_map.px.expand() + sp.O(x**O_n)+sp.O(px**O_n)).removeO()
    x_map.y,x_map.py,x_map.zeta,x_map.delta = 0,0,0,0


print(40*'_')
display(x_map.x)
display(x_map.px)
```

```python
class OneTurnMap:
    def __init__(self, x_map,px_map):
        self.x_map  = sp.lambdify((x,px),x_map)
        self.px_map  = sp.lambdify((x,px),px_map)

    def iterate(self,x0,px0,n_turns):
        x,px = np.nan*np.ones(n_turns),np.nan*np.ones(n_turns)
        x[0],px[0] = x0,px0
        for ii in range(n_turns-1):
            x[ii+1],px[ii+1] = self.x_map(x[ii],px[ii]),self.px_map(x[ii],px[ii])
        return x,px
    



for O_map in [3,4,5,6,7,8]:


    testMap = OneTurnMap(   (x_map.x.expand() + sp.O(x**O_map)+sp.O(px**O_map)).removeO(),
                            (x_map.px.expand() + sp.O(x**O_map)+sp.O(px**O_map)).removeO())
    
    # Plotting resulting phase space

    plt.figure()
    plt.suptitle(f'Truncation order: {O_map}')
    particles,coordinates = generate_particles()
    
    plane = 'x'

    for part_x,part_px in zip(particles.x,particles.px):
        _x,_px =  testMap.iterate(part_x,part_px,1000)

        plt.plot(_x,_px,'o',color='k',alpha=0.5,ms=1,label='space')
            
        

    plt.axis('equal')
    plt.xlabel(f'{plane}_n')
    plt.ylabel(f'p{plane}_n')

    plt.xlim([-2e-4,2e-4])
    plt.ylim([-2e-4,2e-4])


```

```python
import sympy as sp
import numpy as np
sp.init_printing(use_unicode=True,fontsize=24)

from collections import OrderedDict


class Invariant():
    def __init__(self, q_map,p_map,q,p):
        self.C                 = sp.IndexedBase('C')
        self.eps               = sp.symbols('epsilon',real=True)

        self.q,self.p          = q,p
        
        self.q_map             = (1/self.eps*q_map.subs({self.q:self.eps*self.q,
                                                        self.p:self.eps*self.p},simultaneous=True)).expand()
                                                
        self.p_map             = (1/self.eps*p_map.subs({self.q:self.eps*self.q,
                                                        self.p:self.eps*self.p},simultaneous=True)).expand()

        # self.q_map = self.q_map - q*self.q_map.as_coefficients_dict(q,p)[q] -  p*self.q_map.as_coefficients_dict(q,p)[p]
        # self.p_map = self.p_map - q*self.p_map.as_coefficients_dict(q,p)[q] -  p*self.p_map.as_coefficients_dict(q,p)[p]

        # self.q_map = self.q_map/self.eps
        # self.p_map = self.p_map/self.eps

        self.max_order = 5
        self.K0 = self.make_K([(2,0),(1,1),(0,2)])
        self.K1 = self.make_K([(3,0),(2,1),(1,2),(0,3)])
        self.K2 = self.make_K([(4,0),(3,1),(2,2),(1,3),(0,4)])
        self.K3 = self.make_K([(5,0),(4,1),(3,2),(2,3),(1,4),(0,5)])
        self.K4 = self.make_K([(6,0),(5,1),(4,2),(3,3),(2,4),(1,5),(0,6)])
        self.K5 = self.make_K([(7,0),(6,1),(5,2),(4,3),(3,4),(2,5),(1,6),(0,7)])
        
        

    def truncate(self,expr,order):
        return sum(coeff*self.q**pwr[0]*self.p**pwr[1] for pwr,coeff in sp.Poly(expr,self.q,self.p).terms() if pwr[0]+pwr[1]<=order)

    def make_K(self,powers):
        expr   = sum(self.C[pwr[0],pwr[1]]*self.q**pwr[0]*self.p**pwr[1] for pwr in powers)
        return expr

    def Kofn(self,order):
        return sum((self.eps**n)*self.__getattribute__(f'K{n}') for n in range(order+1))

    def residue(self,order):
        K =  self.Kofn(order)
        return (K.subs({self.q:self.truncate(self.q_map,order+1),
                        self.p:self.truncate(self.p_map,order+1)},simultaneous=True) - K).expand()


    def solve_C(self,order,verbose=False,auto_update=True):       
        to_cancel = self.residue(order).as_coefficients_dict(self.eps)[self.eps**order]
        eq_list = list(to_cancel.as_coefficients_dict(self.p,self.q).values())

        to_sub = sp.solve(eq_list,[self.C[i,j] for i in range(5) for j in range(5)])

        if verbose:
            print(60*'-')
            for key,item in to_cancel.as_coefficients_dict(self.p,self.q).items():
                display((sp.Matrix([key]),item))
        
        if auto_update:
            self.update(to_sub,verbose=verbose)
        elif verbose:
            display(to_sub)
        
        if verbose:
            print(60*'-')




    def update(self,to_update,verbose=False):
        if verbose:
            display(to_update)

        for n in range(self.max_order+1):
            # updating the expression
            self.__setattr__(f'K{n}', self.__getattribute__(f'K{n}').subs(to_update, simultaneous=True))


# myInv = Invariant(x_map.x,x_map.px,x,px)
test = Invariant(x_map.x,x_map.px,x,px)
```

```python
def plot_invariant(expr,q,p,eval_at = None,**kwargs):
    K = sp.lambdify((q,p),expr)

    

    if eval_at is not None:
        q_list,p_list = eval_at
        q_list = np.array(q_list)
        p_list = np.array(p_list)

        levels = K(q_list,p_list)
        
        kwargs.update({'levels':np.sort(levels)})

        multiplier = 1
        window = ((multiplier *q_list,-multiplier *q_list,multiplier *p_list,-multiplier *p_list))
        Q,P = np.meshgrid(  np.linspace(np.min(window),np.max(window),500),
                            np.linspace(np.min(window),np.max(window),500))

    else:
        Q,P = np.meshgrid(  np.linspace(-10,10,10000),
                            np.linspace(-10,10,10000))
    plt.contour(Q,P,K(Q,P),**kwargs)


```
